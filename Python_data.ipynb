````python
from wordcloud import WordCloud
import csv
import numpy as np
import PIL 
from collections import Counter

fp = open('/apps.csv', 'r') #The Android App Market on Google Play

fr = csv.reader(fp)

lines = []

for line in fr:
  if len(line) > 0 :
    lines.append(line[1]) #카테고리열만 저장

dict_data = Counter(lines) #카테고리가 몇번씩 나오는지 딕셔너리에 저장
print(dict_data)

img = PIL.Image.open('/abc.png')
img_array = np.array(img)

wc = WordCloud(background_color='white', mask= img_array)
wc = wc.generate_from_frequencies(dict_data)
wc.to_file('/imgWrite.png')  # 워드클라우드가 저장될 파일이름(png)을 넣어주세요
````

####데이터 추출 방법
1. 행번호(iloc)로 추출 - iloc[n]
2. 슬라이싱으로 여러행 추출 - iloc[n:n]
3. 행번호로 여러행 추출 - iloc[[n,n1,n2]]  
4. 특정행, 특정열 - iloc[n][n]  
5. 슬라이싱으로 여러행, 여러열추출 iloc[n:n, n:n]  
6. 슬라이싱 전체 행의 특정 열 iloc[:,n]  
7. 슬라이싱 전체 행의 여러 열 iloc[:,[n, n1, n2]]

````python
#행번호(iloc)로 추출 - iloc[n]

df.iloc[0]
````

````python
#슬라이싱으로 여러행 추출 - iloc[n:n]

df.iloc[1:3]
````

````python
#행번호로 여러행 추출 - iloc[[n,n1,n2]] 

df.iloc[[1,3,5]]
````

````python
#특정행의 특정열 - iloc[n][n]  

df.iloc[2][1] #emma의 국어 점수
````

````python
#학생별 평균 점수 출력

print('이름 \t 합계 \t 평균')

for i in range(len(df)): #df의 행의 개수만큼 반복 --> 각 학생을 구한다
  name = df.iloc[i][0] # i번재 행에서 이름만 가져온다
  total = df.iloc[i][['국어', '영어', '수학']].sum() # i번재 행에서 국어, 영어, 수학을 가져와 더한다.
  avg = df.iloc[i][['국어', '영어', '수학']].mean() # = total / 3
  print(f'{name} \t {total} \t {avg}')
````

````python
#과목별 총점, 평균 구하기

subj = []
total = []
avg = []

for i in range(1, df.shape[1]):
  subj.append(df.columns[i])
  total.append(df.iloc[:,i].sum())
  avg.append(df.iloc[:,i].mean())

a = pd.DataFrame({
    '과목명': subj,
    '총점': total,
    '평균': avg
})

a
````

````python
import matplotlib.pyplot as plt

x=['a','b','c','d']
y=[10,20,30,40]
````

````python
#막대그래프

plt.bar(x, y)
plt.show()
````

````python
#겹친그래프

#그래프 제목
plt.title('sampleGraph')

#그래프 색깔, 레이블
plt.bar(x, y, color='yellow', label = 'bar')
plt.plot(x, y, color='red', label = 'plot')
plt.scatter(x, y, color='blue', label = 'scatter')
plt.legend() #레이블을 보여준다
plt.grid() #그리드 표현시 사용
plt.show()
````

####학생별 점수 파일을 가지고 그래프 그려보기

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/scores.csv') #경로 복사하여 넣기!
````

````python
#이름과 국어 점수만 가지고 그래프 그리기

plt.title('student_score')

x_name = df['이름']
y_kor = df['국어']

plt.bar(x_name, y_kor, color='blue', label = 'Korean')
plt.legend()
plt.grid()
plt.show()
````

````python
#이름과 국어, 영어, 수학 점수 표현하기

#3과목의 x축 차이는 width(0.2)만큼
#국어 : x_index - 0.2
#영어 : x_index
#수학 : x_index + 0.2 

x = df['이름']
y_kor = df['국어']
y_eng = df['영어']
y_math = df['수학']

x_index = range(len(df))
x_kor = list(map(lambda x : x - 0.2, x_index)) #전체 x_kor를 기준보다 앞에 그려라! 왼쪽
x_math = list(map(lambda x : x + 0.2, x_index)) #전체 x_math를 기준보다 뒤에 그려라! 오른쪽

plt.figure(figsize=(10,5))
plt.bar(x_kor,y_kor, color='green',width=0.2)
plt.bar(x_index,y_eng, color='red',width=0.2)
plt.bar(x_math,y_math, color='blue',width=0.2)
plt.title('Midterm exam')
plt.xlabel('name')
plt.ylabel('score')
plt.grid()
plt.show()
````
