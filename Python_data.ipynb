```python
from wordcloud import WordCloud
import csv
import numpy as np
import PIL 
from collections import Counter

fp = open('/apps.csv', 'r') #The Android App Market on Google Play

fr = csv.reader(fp)

lines = []

for line in fr:
  if len(line) > 0 :
    lines.append(line[1]) #카테고리열만 저장

dict_data = Counter(lines) #카테고리가 몇번씩 나오는지 딕셔너리에 저장
print(dict_data)

img = PIL.Image.open('/abc.png')
img_array = np.array(img)

wc = WordCloud(background_color='white', mask= img_array)
wc = wc.generate_from_frequencies(dict_data)
wc.to_file('/imgWrite.png')  # 워드클라우드가 저장될 파일이름(png)을 넣어주세요
```

```python
#한글파일로 만들어보기 txt --> 현재 첫번째 줄만 인식함 why??????????

from wordcloud import WordCloud
import numpy as np
import PIL 
from collections import Counter

fp = open('/babyshark.txt', 'r', encoding='utf-8') #아기상어 가사 텍스트 파일

lines = []

for line in fp.readline():
  if len(line) > 0 :
    lines.append(line.replace('\n', '')) #가사 속 \n 제거

dict_data = Counter(lines)
print(dict_data)

img = PIL.Image.open('/abc.png') #참고할 이미지
img_array = np.array(img)

wc = WordCloud(font_path = 'NanumGothic.ttf', background_color='white', mask= img_array)
wc = wc.generate_from_frequencies(dict_data)
wc.to_file('/imgBabyShark.png') # 워드클라우드가 저장될 파일이름(png) 넣어주세요
```

####데이터 추출 방법
1. 행번호(iloc)로 추출 - iloc[n]
2. 슬라이싱으로 여러행 추출 - iloc[n:n]
3. 행번호로 여러행 추출 - iloc[[n,n1,n2]]  
4. 특정행, 특정열 - iloc[n][n]  
5. 슬라이싱으로 여러행, 여러열추출 iloc[n:n, n:n]  
6. 슬라이싱 전체 행의 특정 열 iloc[:,n]  
7. 슬라이싱 전체 행의 여러 열 iloc[:,[n, n1, n2]]

```python
#행번호(iloc)로 추출 - iloc[n]

df.iloc[0]
```

```python
#슬라이싱으로 여러행 추출 - iloc[n:n]

df.iloc[1:3]
```

```python
#행번호로 여러행 추출 - iloc[[n,n1,n2]] 

df.iloc[[1,3,5]]
```

```python
#특정행의 특정열 - iloc[n][n]  

df.iloc[2][1] #emma의 국어 점수
```

```python
#슬라이싱으로 여러행, 여러열추출 iloc[n:n, n:n]

df.iloc[1:4, 0:2] #1~3 행, 0~1 열을 가져온다
```

```python
#슬라이싱 전체 행의 특정 열 iloc[:,n]

df.iloc[:, 0] #0번째 열의 전체 행을 가져온다
```

```python
#슬라이싱 전체 행의 여러 열 iloc[:,[n, n1, n2]]

df.iloc[:, [0, 2, 3]]
```
