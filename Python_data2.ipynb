####춘천 기온 그래프

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

x = df['년']
y = df['평균']

plt.title('춘천 평균 기온')

plt.plot(x,y,label='평균')
plt.grid()
plt.legend()
plt.show()
```

```python
#최고기온 그래프 구하기

x = df['년']
y = df['최고']

plt.title('춘천 최고 기온')

plt.plot(x,y,color='red',label='최고기온')
plt.grid()
plt.legend()
plt.show()
```

```python
#최저기온 그래프 구하기
plt.rcParams['axes.unicode_minus'] = False # (-) 영하 값 표현

x = df['년']
y = df['최저']

plt.title('춘천 최저 기온')

plt.plot(x,y,color='green',label='최저기온')
plt.grid()
plt.legend()
plt.show()
```

####내 생일 기후 변천사

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/knu_temp_day.csv', encoding='utf-8')
df = df.drop(columns = ['지점'])
df.columns = ['일자', '평균', '최저', '최고']
df.dropna(inplace=True)
df.isnull().sum()

df['생일'] = df['일자'].str[5:]
df['생년'] = df['일자'].str[:4]
df1 = df[df['생일'] == '04-19']

x = df1['생년']
a = df1['평균']
b = df1['최고']
c = df1['최저']

plt.figure(figsize=(20,5))
plt.title('내 생일 기후 변천사')
plt.plot(x,a,color='pink')
plt.plot(x,b,color='yellow')
plt.plot(x,c,color='blue')
plt.show()
```
